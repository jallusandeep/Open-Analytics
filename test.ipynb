```python
import requests
import json
from urllib.parse import quote

ANALYTICS_TOKEN = "eyJ0eXAiOiJKV1QiLCJrZXlfaWQiOiJza192MS4wIiwiYWxnIjoiSFMyNTYifQ.eyJzdWIiOiI1UkNQUDIiLCJqdGkiOiI2YTFiYjg5NzliYzM4NDVkYmRhZjdjYjYiLCJpc011bHRpQ2xpZW50IjpmYWxzZSwiaXNQbHVzUGxhbiI6dHJ1ZSwiaXNFeHRlbmRlZCI6dHJ1ZSwiaWF0IjoxNzgwMjAxNjIzLCJpc3MiOiJ1ZGFwaS1nYXRld2F5LXNlcnZpY2UiLCJleHAiOjE4MTE4MDA4MDB9.OQW9uTW2mRTb8ZgnMYszcjKApikWCDOP3PClg_4E5pM"

BASE = "https://api.upstox.com"

headers = {
    "Authorization": f"Bearer {ANALYTICS_TOKEN}",
    "Accept": "application/json"
}

def get_expiries(symbol="NIFTY50", exchange_segment="NSE_INDEX"):
    url = f"{BASE}/v2/expired-instruments/expiries"
    params = {
        "symbol": symbol,
        "exchange_segment": exchange_segment
    }
    r = requests.get(url, headers=headers, params=params, timeout=30)
    print("Expiries status:", r.status_code)
    print("Expiries raw:", r.text)        # print full response
    r.raise_for_status()
    return r.json()

def get_expired_option_contracts(
    underlying_symbol="NIFTY50",
    exchange_segment="NSE_INDEX",
    expiry_date="2025-04-24"
):
    url = f"{BASE}/v2/expired-instruments/option/contract"
    params = {
        "symbol": underlying_symbol,
        "exchange_segment": exchange_segment,
        "expiry_date": expiry_date
    }
    r = requests.get(url, headers=headers, params=params, timeout=30)
    print("Expired option contracts status:", r.status_code)
    print("Expired option contracts raw:", r.text)
    r.raise_for_status()
    return r.json()

def get_expired_historical_candles(
    expired_instrument_key,
    interval="1minute",
    from_date="2025-04-01",
    to_date="2025-04-24"
):
    encoded_key = quote(expired_instrument_key, safe="")
    url = (
        f"{BASE}/v2/expired-instruments/historical-candle/"
        f"{encoded_key}/{interval}/{to_date}/{from_date}"
    )
    r = requests.get(url, headers=headers, timeout=30)
    print("Expired historical candles URL:", url)
    print("Expired historical candles status:", r.status_code)
    print("Expired historical candles raw:", r.text)
    r.raise_for_status()
    return r.json()

# ---- Example flow ----

# 1) Get expiries for NIFTY50
expiries = get_expiries()
print("Parsed expiries JSON:", json.dumps(expiries, indent=2))

# Pick one expiry_date from the response manually or programmatically
expiry_date = "2025-04-24"  # example

# 2) Get expired option contracts for that expiry
contracts = get_expired_option_contracts(expiry_date=expiry_date)
print("Parsed contracts JSON:", json.dumps(contracts, indent=2))

# Pick one expired_instrument_key from contracts
expired_key = "NSE_FO|73507|24-04-2025"  # replace with a real key from contracts

# 3) Get historical candles for that expired contract
candles = get_expired_historical_candles(
    expired_instrument_key=expired_key,
    interval="1minute",
    from_date="2025-04-01",
    to_date="2025-04-24"
)
print("Parsed candles JSON:", json.dumps(candles, indent=3))
```

```python
import time
import pandas as pd
from io import StringIO

def get_ipo_gmp_df(url="https://ipowatch.in/ipo-grey-market-premium-latest-ipo-gmp/"):
    target_cols = [
        "IPO Name", "IPO GMP", "Trend", "Price Band",
        "Est. Listing", "Date", "Type", "Status", "Last Updated"
    ]

    # ---------- Method 1: pandas.read_html ----------
    try:
        tables = pd.read_html(url)

        for df in tables:
            cols = [str(c).strip() for c in df.columns]
            if "IPO Name" in cols and "IPO GMP" in cols:
                df = df.copy()
                df.columns = cols
                return df

    except Exception as e:
        print("read_html failed, trying Selenium fallback...")
        print("Reason:", e)

    # ---------- Method 2: Selenium fallback ----------
    from selenium import webdriver
    from selenium.webdriver.common.by import By
    from selenium.webdriver.support.ui import WebDriverWait
    from selenium.webdriver.support import expected_conditions as EC

    options = webdriver.ChromeOptions()
    options.add_argument("--headless=new")
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--window-size=1920,1080")

    driver = webdriver.Chrome(options=options)
    wait = WebDriverWait(driver, 25)

    try:
        driver.get(url)

        wait.until(EC.presence_of_all_elements_located((By.TAG_NAME, "table")))
        time.sleep(2)

        tables = driver.find_elements(By.TAG_NAME, "table")
        html_table = None

        for table in tables:
            text = table.text.strip()
            if (
                "IPO Name" in text and
                "IPO GMP" in text and
                "Price Band" in text and
                "Last Updated" in text
            ):
                html_table = table.get_attribute("outerHTML")
                break

        if html_table is None:
            raise Exception("Target table not found even with Selenium")

        df = pd.read_html(StringIO(html_table))[0]
        df.columns = [str(c).strip() for c in df.columns]

        return df

    finally:
        driver.quit()


# ------------------ Usage ------------------
url = "https://ipowatch.in/ipo-grey-market-premium-latest-ipo-gmp/"
df = get_ipo_gmp_df(url)

print(df)
```
